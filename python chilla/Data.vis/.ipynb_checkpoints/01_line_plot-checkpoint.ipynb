# import libraries
- Seaborn automatically initials these libraries
- numpy
- scipy
- pandas
- matplotlib

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load dataset
phool = sns.load_dataset("iris")


#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=phool)
plt.show()
```

## Adding titles

```python
#load dataset
phool = sns.load_dataset("iris")


#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=phool)
plt.title("Phoolon ka plot")
plt.show()
```

## Adding Limits

```python
#load dataset
phool = sns.load_dataset("iris")


#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=phool)
plt.title("Phoolon ka plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## Set Styles
- dark grid
- whitegrid
- dark grid
- white
- ticks

```python
#load dataset
phool = sns.load_dataset("iris")

#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=phool)
plt.title("Phoolon ka plot")
sns.set_style("dark")
sns.set_style(style=None,rc=None)
plt.show()
```

## size of Figure

```python
#load dataset
phool = sns.load_dataset("iris")
sns.set_style(style=None,rc=None)
# change figure
plt.figure(figsize=(8,6))

#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=phool)
plt.title("Phoolon ka plot")
sns.set_style("whitegrid")
plt.show()
```
